```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
from datetime import date
import psycopg2
import pandas as pd
from datetime import timedelta 
```

```python
#EPIC PASS
for x in range (0, len(epic_href)):
#for x in range (1, 3):
    url = f"{base_url}{epic_href[x]}"
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')  
    browser = webdriver.Chrome(options=options)
    browser.get(url)
    html = browser.page_source
    time.sleep(10)
    soup = BeautifulSoup(html, "lxml")
    today_date = date.today()
    today_date = today_date - timedelta(days=1)
    
    #Parse text
    print('-----------------------')
    print(f"Result #{x+1} out of 49")
    name = soup.find('span', class_="resort_name").text
    location_ = epic_href[x].split("/")
    location= location_[1]
    today_ = soup.findAll('div' , class_="predicted_snowfall")
    today = today_[5].find('div', class_="bluePill").text
    try:
        upper = soup.find('li', class_="elevation upper").find('div', class_="bluePill").text
    except:
        upper = str("N/A")
    try:
        mid = soup.find('li', class_="elevation middle").find('div', class_="bluePill").text
    except:
        mid = str("N/A")
    try:
        lower = soup.find('li', class_="elevation lower").find('div', class_="bluePill").text
    except:
        lower = str("N/A")
    #Format
    try:
        upper = upper.replace('"', '')
        upper = int(upper)
    except:
        upper = str("N/A")
    try:
        mid = mid.replace('"', '')
        mid = int(mid)
    except:
        mid = str("N/A")
    try:
        lower = lower.replace('"', '')
        lower = int(lower)
    except:
        lower = str("N/A")
    try:
        today = today.replace('"', '')
        today = int(today)
    except:
        today = str("N/A")
    pass_ = str("EPIC")
    try:
        location = location.replace('-', ' ')
    except:
        location = location
    location = location.title()
    try:
        location = location.replace(' ', '_')
    except:
        location = location
    if location =="Trentino":
        country = "Italy"
    elif location =="British_Columbia":
        country = "Canada"
    elif location =="Washington":
        country = "United_States"
    elif location =="Northern_Alps":
        country = "France"
    elif location =="California":
        country = "United_States"
    elif location =="Valais":
        country = "Switzerland"
    elif location =="Alberta":
        country = "Canada"
    elif location =="Vermont":
        country = "United_States"
    elif location =="Idaho":
        country = "United_States"
    elif location =="Utah":
        country = "United_States"
    elif location =="Colorado":
        country = "United_States"
    elif location =="Michigan":
        country = "United_States"
    elif location =="Minnesota":
        country = "United_States"
    elif location =="Wisconsin":
        country = "United_States"
    elif location =="New_Hampshire":
        country = "United_States"
    elif location =="Quebec":
        country = "Canada"
    elif location =="Australia":
        country = "Australia"
    elif location =="Wyoming":
        country = "United_States"
    elif location =="Oregon":
        country = "United_States"
    elif location =="West_Virginia":
        country = "United_States"
    elif location =="Montana":
        country = "United_States"
    elif location =="New_York":
        country = "United_States"
    elif location =="Ontario":
        country = "Canada"
    elif location =="Maine":
        country = "United_States"
    elif location =="New_Mexico":
        country = "United_States"
    else:
        country = "N/A"
    
    #Print Results
    print(today_date)
    print(f"Resort name: {name}")
    print(f"Region: {location}")
    print(f"Country: {country}")
    print(f"Pass: {pass_}")
    print(f"Today's snowfall: {today}")
    print(f"Upper depth: {upper}")
    print(f"Mid depth: {mid}")
    print(f"Lower depth: {lower}")
    browser.quit()

    #Save results to Dataframe
    snowfall_df = snowfall_df.append({'Date': today_date, 'Resort_Name': name,'Region': location,'Country': country, 'Pass':pass_,\
    'Snowfall':today,'Lower_Depth':lower, 'Middle_Depth': mid,'Upper_Depth': upper}, ignore_index=True)
```
